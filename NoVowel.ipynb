```python
x = "A vowel is a particular kind of speech sound made by changing the shape of the upper vocal tract, or the area in the mouth above the tongue. In English it is important to know that there is a difference between a vowel sound and a letter in the alphabet"
vowel = ['a', 'e', 'i', 'o', 'u','A','E','I','O','U']

stri = x.split()

for word in stri:
    present = False
    for j in word:
        if j in vowel:
            present = True
    if present == False:
        print(word, end=" ")
        
```
